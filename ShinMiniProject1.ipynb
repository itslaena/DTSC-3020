## Part 2 — The Receipt & Loyalty Stars  *(25 points)*

A customer orders **3 lattes** ($4.50 each) and **2 muffins** ($3.25 each). Texas sales tax is **8.25%**. The shop's loyalty program gives **1 star for every whole $5 spent** (based on the total).

### What to do
1. Store the four order values in variables: `latte_price`, `latte_qty`, `muffin_price`, `muffin_qty`.
2. Calculate, using **arithmetic operators**:
   - `subtotal` — price × quantity for both items, added together
   - `tax` — `subtotal * 8.25 / 100`
   - `total` — subtotal plus tax
3. Print the receipt using **f-strings with 2 decimal places**, like this:

   ```
   Subtotal: $20.00
   Tax (8.25%): $1.65
   Total: $21.65
   ```

   > 💡 **Money formatting recipe:** `f"${subtotal:.2f}"` prints a float with exactly 2 decimals.
4. Loyalty stars: first convert the total to whole dollars with `int(total)` (store it in `total_whole`). Then use **floor division and modulo**:
   - `stars_earned = total_whole // 5`
   - `toward_next = total_whole % 5`
   
   Print both with labels, e.g. `Stars earned: 4` and `Dollars toward next star: 1`.
5. The customer brought a **$25.00 budget**. Compute and print how much money they have left, formatted as money.
6. Add a **comment** answering: why do `//` and `%` together "split" the dollars into stars and leftover? (One sentence — hint: `21 = 4 * 5 + 1`.)


````python
# Hyelim Laena Shin — Part 2

latte_price = 4.50
latte_qty = 3
muffin_price = 3.25
muffin_qty = 2

subtotal = latte_price * latte_qty + muffin_price * muffin_qty
tax = subtotal * 8.25 / 100
total = subtotal + tax

print(f"Subtotal: ${subtotal:.2f}")
print(f"   Tax (8.25%): ${tax:.2f}")
print(f"Total: ${total:.2f}")

total_whole = int(total)
stars_earned = total_whole // 5
toward_next = total_whole % 5
print("Stars earned:", stars_earned)
print("Dollars toward next star:", toward_next)

budget = 25.00
remaining = budget - total
print(f"Change from $25.00 budget: ${remaining:.2f}")

# // and % together split the total into stars and leftover because floor division gives
# how many complete groups of 5 fit into the total, and modulo gives what's left over —
# e.g., 20 = 4 * 5 + 0, so 4 stars and $0 remaining.
````
